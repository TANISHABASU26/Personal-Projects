Undersampling, building a sample data set from the original dataset containing similar distribution from normal transactions and fraudelent transaction.Taking 492 samples, randomly from the fraudelent transactions attach to fraud data set.

```python
legit_sample =legit_data.sample(n=492)
```

```python
#concatenate two data frames
```

```python
new = pd.concat([legit_sample, fraud_data],axis=0) #if it was 1, then the values would be added row wise, if 0 then column wise which we want.
```

```python
new.head()
```

```python
new.tail()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
#split the data
```

```python
X = new.drop(columns='Class',axis=1)
Y =new['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split into training and testing data.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size= 0.2 ,stratify = Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
```

Model Training:

logistic regression model.

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#TRAINNG THE LOGISTIC REGRESSION MODEL WITH TRAINING DATA:
```

```python
#accuracy score
```

```python
X_train_prediction = model.predict(X_train)
trainingaccuracy = accuracy_score(X_train_prediction,Y_train)
print('Training data accuracy is :',trainingaccuracy)
```

```python
X_test_prediction = model.predict(X_test)
testdataaccuracy = accuracy_score(X_test_prediction,Y_test)
print('Test data accuracy is :',testdataaccuracy)
```
